```python
import math

points = {
    'P1': [0.1, 0.6],
    'P2': [0.15, 0.71],
    'P3': [0.08, 0.9],
    'P4': [0.16, 0.85],
    'P5': [0.2, 0.3],
    'P6': [0.25, 0.5],
    'P7': [0.24, 0.1],
    'P8': [0.3, 0.2]
}

m1 = points['P1']
m2 = points['P8']

def distance(p, q):
  return math.sqrt(((p[0] - q[0]) ** 2) + ((p[1] - q[1]) ** 2) )

cluster1 = []
cluster2 = []

for name , coord in points.items():
  d1 = distance(coord, m1)
  d2 = distance(coord, m2)

  if d1 < d2 :
    cluster1.append(name)
  else :
    cluster2.append(name)

print(f"\ncluster 1 members:: C1 = {cluster1}")
print(f"cluster 2 members:: C2 = {cluster2}\n")

def mean_of_clusters(cluster):
  x_sum = 0
  y_sum = 0

  for name in cluster:
    x_sum += points[name][0]
    y_sum += points[name][1]

  return [x_sum/len(cluster), y_sum/len(cluster)]

mean_m1 = mean_of_clusters(cluster1)
mean_m2 = mean_of_clusters(cluster2)

print(f"\nThe updated Centroids are : m1 = {mean_m1[0]:.3f}, {mean_m1[1]:.3f}\n")
print(f"The updated Centroids are : m2 = {mean_m2[0]:.3f}, {mean_m2[1]:.3f}\n")

print("The P6 belong to: ", "C1" if 'P6' in cluster1 else "C2")
print("Population of cluster around m2:", len(cluster2))
```
