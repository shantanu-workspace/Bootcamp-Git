```python
# üß† Step 1: Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# üìÇ Step 2: Load Dataset
# Make sure 'test.csv' is in the same directory or provide full path
df = pd.read_csv('train.csv')

# Display first few rows
print(df.head())

# üîç Step 3: Separate Features (X) and Target (y)
X = df.drop('price_range', axis=1)
y = df['price_range']

# Check shape of data
print("Feature shape:", X.shape)
print("Target shape:", y.shape)

# ‚úÇÔ∏è Step 4: Split Data into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
print("Train set size:", X_train.shape)
print("Test set size:", X_test.shape)

# ‚öôÔ∏è Step 5: Scale Features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ü§ñ Step 6: Train and Compare SVM Models with Different Kernels
kernels = ['linear', 'rbf', 'poly']
results = {}

for kernel in kernels:
    print(f"\nüî∏ Training SVM with {kernel} kernel...")
    svm = SVC(kernel=kernel, random_state=42)
    svm.fit(X_train_scaled, y_train)
    y_pred = svm.predict(X_test_scaled)

    acc = accuracy_score(y_test, y_pred)
    results[kernel] = acc

    print(f"Accuracy ({kernel}): {acc:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    # üìä Step 7: Summary of Accuracies
print("\n‚úÖ Kernel-wise Accuracy Summary:")
for k, v in results.items():
    print(f"{k.capitalize()} Kernel: {v:.4f}")
```
