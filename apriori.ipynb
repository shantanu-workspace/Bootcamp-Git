```python
import pandas as pd
from itertools import combinations

df = pd.read_csv("Order1.csv")

transactions = df.groupby(['Member_number' , 'Date'])['itemDescription'].apply(list)

transaction_list = []
for t in transactions:
  transaction_list.append(t)

item_count = {}
for basket in transaction_list:
  for item in basket:
    if item in item_count:
      item_count[item] += 1
    else:
      item_count[item] = 1

min_support = 0.01
total_transactions = len(transaction_list)

#made the first frequent list
L1 = {}
for item, count in item_count.items():
  support = count / total_transactions
  if support >= min_support:
    L1[item] = count

print("\n\n=== Frequent 1-itemsets (L1) ===\n")
for item , count in L1.items():
  print(f"{item}: support={count / total_transactions:3f}")
print("\n")



#halfway



pair_count = {}
for basket in transaction_list:
  filtered_pairs = []
  for item in basket:
    if item in L1:
      filtered_pairs.append(item)

  for pair in combinations(filtered_pairs, 2):
    pair = tuple(sorted(pair))
    if pair in pair_count:
      pair_count[pair] += 1
    else:
      pair_count[pair] = 1


#made the second frequent list
L2 = {}
for pair, count in pair_count.items():
  support = count / total_transactions
  if support >= min_support:
    L2[pair] = count

print("\n\n=== Frequent 2-itemsets (L2) ===")
for pair , count in L2.items():
  print(f"{pair} : support = {count / total_transactions:.3f}")
print(f"\n")

print(f"\n\n=============Association Rules=============\n")
for (itemA, itemB), countAB in L2.items():
  #confidence A -> B = support(a -> b) / support(a)
  confidence_A_to_B = countAB / item_count[itemA]
  confidence_B_to_A = countAB / item_count[itemB]

  print(f"{itemA} -> {itemB} : support = {confidence_A_to_B:.2f}")
  print(f"{itemB} -> {itemA} : support = {confidence_B_to_A:.2f}")

print("Exiting...\n")

```

```python
import pandas as pd
from itertools import combinations

# STEP 1: Load dataset (no header)
df = pd.read_csv("Order2.csv", header=None)

# STEP 2: Convert rows to transactions (remove NaN values)
transaction_list = []
for i in range(len(df)):
    basket = []
    for item in df.iloc[i]:
        if pd.notna(item):
            basket.append(item)
    transaction_list.append(basket)

# STEP 3: Count frequency of each item (for L1)
item_count = {}
for basket in transaction_list:
    for item in basket:
        if item in item_count:
            item_count[item] += 1
        else:
            item_count[item] = 1

# STEP 4: Filter by minimum support
min_support = 0.01
total_transactions = len(transaction_list)

L1 = {}
for item, count in item_count.items():
    support = count / total_transactions
    if support >= min_support:
        L1[item] = count

print("\n=== Frequent 1-Itemsets (L1) ===\n")
for item, count in L1.items():
    print(f"{item}: support = {count / total_transactions:.3f}")


# STEP 5: Generate frequent 2-itemsets (L2)
pair_count = {}
for basket in transaction_list:
    filtered_items = []
    for item in basket:
        if item in L1:
            filtered_items.append(item)

    for pair in combinations(filtered_items, 2):
        pair = tuple(sorted(pair))
        if pair in pair_count:
            pair_count[pair] += 1
        else:
            pair_count[pair] = 1

L2 = {}
for pair, count in pair_count.items():
    support = count / total_transactions
    if support >= min_support:
        L2[pair] = count

print("\n\n=== Frequent 2-Itemsets (L2) ===\n")
for pair, count in L2.items():
    print(f"{pair}: support = {count / total_transactions:.3f}")


# STEP 6: Generate association rules (Confidence)
print("\n\n=== Association Rules ===\n")
for (itemA, itemB), countAB in L2.items():
    confidence_A_to_B = countAB / item_count[itemA]
    confidence_B_to_A = countAB / item_count[itemB]

    print(f"{itemA} -> {itemB}, confidence = {confidence_A_to_B:.2f}")
    print(f"{itemB} -> {itemA}, confidence = {confidence_B_to_A:.2f}")
    print("-----")
```
